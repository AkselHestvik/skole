## Lesing og plotting av data

```python
from pylab import *

data = loadtxt("sunspots.txt", skiprows = 1, delimiter = ",")

year = data[:,0]
sunactivity = data[:,1]

plot(year, sunactivity)
show()
```

Her bruker du loadtxt som en funskjon, og du skriver inn filnavnet. Det er viktig √• merke seg at for √• bare skrive filnavnet m√• filen du vil bruke ligge i samme plassering som programmet. Skiprows betyr at vi skipper f√∏rste rad. Det gj√∏r jeg her fordi det er tekst. Jeg bruker delimiter for √• si at jeg skiller dataene med et "," tegn

## Regresjon

```python
from pylab import *
from sklearn.metrics import r2_score

T = array([0, 20, 40, 60, 80, 100])
sol_NH3 = array([88.5, 56.0, 34.0, 20.0, 11.0, 7.0])
sol_NaCl = array([35.7, 35.9, 36.4, 37.1, 38.0, 39.2])

#regresjon
n = 2 #grad av funksjonen
reg = polyfit(T, sol_NH3, n)

x = linspace(0, 150, 10000)
y = reg[0]*x**2 + reg[1]*x + reg[2]

#R2 score, alts√• hvor god er grafen
y_R2 = reg[0]*T**2 + reg[1]*T + reg[2]
R2 = r2_score(sol_NH3, y_R2)
print(R2)

scatter(T, sol_NH3, label = "Datapunkter, NH$_3$")
plot(x, y, color = 'red', label='Tilpasset kurve, NH$_3$')
legend()
title("l√∏slighet av salter")
xlabel("Temperatur i $^o$C")
ylabel("L√∏slighet i g/100ml")
grid()
show()
```

Her ser man at R2 scoren er 0.998, som er veldig n√¶rme 1, noe som betyr at grafen passer veldig bra. Regresjon handler jo om √• finne en funksjon som passer bra til noen punkter
